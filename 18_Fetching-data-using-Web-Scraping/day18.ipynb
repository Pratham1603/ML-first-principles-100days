ALL IN ONE FUNCTION FOR PAGE 1 

```python
name=[]
child=[]
reviews=[]
ctype=[]
hq=[]

for i in company:

    name.append(i.find_all('h2')[0].text.strip())
    child.append(i.find_all('div', class_='rating_text rating_text--md')[0].text.strip())
    reviews.append(i.find_all('span', class_='companyCardWrapper__ActionCount')[0].text.strip())
    ctype.append(i.find_all('span', class_='companyCardWrapper__interLinking')[0].text.split('|')[0].strip())
    hq.append(i.find_all('span', class_='companyCardWrapper__interLinking')[0].text.split('|')[1].strip().split()[0])

df =pd.DataFrame({'Name':name,'Rating':child,'Reviews':reviews,'Company Type':ctype,'Headquarters':hq})
```

# creating dataframe for all the pages

```python
final=pd.DataFrame()
for j in range(1,11):   # since all 1000 page take so much amount of time we are fetching only 10 pages
    webpage=requests.get(f'https://www.ambitionbox.com/list-of-companies?page={j}',headers=headers).text
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div',class_='companyCardWrapper')
    
    name=[]
    child=[]
    reviews=[]
    ctype=[]
    hq=[]

    for i in company:

        try:
         name.append(i.find_all('h2')[0].text.strip())
        except:
         name.append(np.nan)

        
        try:
         child.append(i.find_all('div', class_='rating_text rating_text--md')[0].text.strip())
        except:
         child.append(np.nan)

        try:
         reviews.append(i.find_all('span', class_='companyCardWrapper__ActionCount')[0].text.strip())
        except:
         reviews.append(np.nan)

        try:
         ctype.append(i.find_all('span', class_='companyCardWrapper__interLinking')[0].text.split('|')[0].strip())
        except:
         ctype.append(np.nan)

        try:
         hq.append(i.find_all('span', class_='companyCardWrapper__interLinking')[0].text.split('|')[1].strip().split()[0])
        except:
         hq.append(np.nan)

    df = pd.DataFrame({'Name':name,'Rating':child,'Reviews':reviews,'Company Type':ctype,'Headquarters':hq})
    
    final = pd.concat([final, df], ignore_index=True)
```
