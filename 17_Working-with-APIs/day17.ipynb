# Working with APIs

### Converting API into DataFrame

```python
import pandas as pd
import requests
```

Hitting the server with this api 

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
```

```python
response  # 200 - good, 404 - not found, 500 - server error
```

Making tempary dataframe for 1 page of json response outoff 426 pages which I know coz, I checked on website known as jsonviewer for better visualiztion my json file.

Extracting imp cols only

```python
temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
temp_df
```

Create empty dataframe for storing all pages, So we use loop for it

```python
df = pd.DataFrame()
```

```python
df
```

```python
for i in range(1,429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = df.append(temp_df,ignore_index=True)
    
```

```python
df
```

```python
df.shape
```

To export this dataframe in CSV form 

```python
df.to_csv('movies.csv')
```
