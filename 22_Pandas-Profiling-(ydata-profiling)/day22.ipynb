# TOOL : Y-DATA-PROFILING FOR EDA ANALYSIS 

Since panads-profiling is outdated we use ydata-profiling 

It is use for Auto-EDA Analysis ( No manual EDA is needed)

Note : Always use ydata-profiling for EDA to save time and do indepth EDA

```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')   # Titanic dataset
```

```python
df.head()
```

```python
df.info()
```

```python
!pip install ydata-profiling
```

Working : It gives EDA Report as html file ( output.html )

```python
from ydata_profiling import ProfileReport

prof = ProfileReport(df)
prof.to_file("output.html")
```

```python
 # Now check the output.html file in the current directory to see the profiling report.
```
